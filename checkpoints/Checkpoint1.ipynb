```python
#Load reviews.csv into a pd.DataFrame named car_rev
import pandas as pd
import matplotlib.pyplot as plt

#Read and clean data
car_rev = pd.read_csv('/content/drive/MyDrive/porschereviews.csv', on_bad_lines='skip', engine='python')
car_rev = car_rev.dropna()
car_rev = car_rev.drop_duplicates()
clean_time = car_rev['Review_Date'].str.extract(r'(\d{2}/\d{2}/\d{2} \d{2}:\d{2} [AP]M)')[0] #remove "on"
car_rev['Review_Date_dt'] = pd.to_datetime(clean_time, format='%m/%d/%y %H:%M %p') #add formatting for the date column to extract dates in numeric

#Initial View of data size and columns
print("Rows: " + str(car_rev.shape[0]) + " Columns: " + str(car_rev.shape[1]))
print()
print(car_rev.info())
print()
print(car_rev.head(n=5))

#EDA: Column exploration
vehicle_counts = car_rev['Vehicle_Title'].value_counts(ascending=False, normalize=False)
ratings = car_rev['Rating'].value_counts(ascending=False, normalize=False)
review_len = car_rev['Review'].str.len()
year_count = car_rev['Review_Date_dt'].dt.year.value_counts(ascending=False, normalize=False)

print()
print(vehicle_counts)
print()
print(ratings)
print()
print(review_len)
print()
print(year_count)

#A matplotlib for distribution plotting of review length
plt.hist(review_len, bins=20)
plt.title("Porsche Consumer Review Length Distribution")
plt.xlabel("Char Length")
plt.ylabel("Reviews")
plt.show()

#Ratings
plt.bar(ratings.index, ratings.values)
plt.title("Porsche Consumer Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

#Years
plt.bar(year_count.index, year_count.values)
plt.title("Porsche Consumer Rating Over the Years")
plt.xlabel("Year")
plt.ylabel("Review Count")
plt.show()
```

EDA Steps & Observations

For my EDA I started by loading and cleaning the data from the Porsche Consumer Review dataset. There were no empty or unnamed columns, but there were rows with missing values. These were identified as reviews left with no content or review description, just a rating. With the purpose of the project being text mining from the actual review content, these rows were dropped and the size of the data shrunk by about 33%. The remaining rows were analyzed and initial information was collected based on importance. Measures like time and author are not extremely substantial to the data as a whole. However, car model, rating, and review length were values I thought would be insightful enough to view. The final data cleaning involved adding a column for the date in the datetime format. This would allow for extraction later on. I measured frequencies of mentioned car models, ratings and year of review, as well as the distribution of ratings and the length of the reviews. The latter gave insight to potential bias within the consumer review system. The ratings are overwhelmingly high, with a whopping 40% of reviews being a full 5 stars. Keeping in mind that this is a highly acclaimed luxury brand, it does make sense for ratings to be high, especially upon the purchase of said vehicle being received well and enjoyed. However, the amount of positive ratings skew the distribution, leaving room to question bias in self-reporting, possibly filtering reviews to focus on the positive ones, and selection bias for those who have owned cars but maybe not for long, but not those who have had experiences with a Porsche that were worth writing a review about but may not own it.
